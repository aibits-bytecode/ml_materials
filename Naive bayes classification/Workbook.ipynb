### Load Data

```python
import pandas as pd
df = pd.read_csv('./data/titanic.csv')
df.head()
```

```python
df = df[['Pclass','Sex','Age','Fare','Survived']]
df.head()
```

```python
y = df['Survived']
x = df.drop(['Survived'],axis=1)
```

```python
x
```

### Categorical data

```python
df_sex_dummies = pd.get_dummies(df.Sex)
df_sex_dummies
```

```python
x_concat = pd.concat([x,df_sex_dummies], axis=1)
x_concat
```

```python
x = x_concat.drop(['female','Sex'],axis=1)
x
```

### Data preprocessing

```python
print('Pclass : ', str(x['Pclass'].isnull().values.sum()))
print('Age : ', str(x['Age'].isnull().values.sum()))
print('Fare : ', str(x['Fare'].isnull().values.sum()))
print('male : ', str(x['male'].isnull().values.sum()))
```

```python
x.Age = x.Age.fillna(x.Age.median())
print('Age : ', str(x['Age'].isnull().values.sum()))
x.head()
```

### Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x.Fare = scaler.fit_transform(x[['Fare']])
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

### Model creation

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

## Spam Email ðŸš‘

```python
import pandas as pd
df = pd.read_csv('./data/spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.Message,df.spam, test_size=0.2)
```

### Words Counts 

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

### Model creation

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
x_test_count = v.transform(x_test)
model.score(x_test_count, y_test)
```

### Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vect', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```
