# Data preprocessing

### Fillna values

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/homeprices.csv')
df1.head()
```

```python
# In here we are going to use the median for the missing value
import math 
median_bedrooms = math.floor(df1.bedrooms.median())
median_bedrooms
```

```python
df1.bedrooms = df1.bedrooms.fillna(median_bedrooms)
df1.head()
```

### Label encoding <br>
Use integers to represent the string values

Drawback is model would be think monroe is better than west windsor or wiseversa

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_labelEncoder = df
df_labelEncoder.town = le.fit_transform(df_labelEncoder.town)
df_labelEncoder
```

### One hot encoding

```python
df.head()
```

```python
df_town_dummies = pd.get_dummies(df.town)
df_town_dummies
```

```python
df_concat = pd.concat([df,df_town_dummies], axis=1)
df_concat
```

```python
df_concat_last = df_concat.drop(['west windsor','town'], axis=1)
df_concat_last
```

In sklearn it knows of the dummy variable trap and it drops the column automatically<br>

### Split for model creation

### Kfold cross validation

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
lr = LogisticRegression()
rf = RandomForestClassifier()
```

```python
cross_val_score(lr,digits.data,digits.target)
```

```python
cross_val_score(rf,digits.data,digits.target,cv=3)
```
